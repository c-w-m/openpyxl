## Inserting rows and columns


You can insert rows or columns using the relevant worksheet methods:

    * `openpyxl.worksheet.worksheet.Worksheet.insert_rows`
    * `openpyxl.worksheet.worksheet.Worksheet.insert_cols`
    * `openpyxl.worksheet.worksheet.Worksheet.delete_rows`
    * `openpyxl.worksheet.worksheet.Worksheet.delete_cols`

The default is one row or column. For example to insert a row at 3 (before
the existing row 3)::

```python
from openpyxl import load_workbook

wb = load_workbook(filename='large_file_before.xlsx', read_only=False)
ws = wb['big_data']

for row in ws.rows:
    for cell in row:
        print(cell.value)
```

```python
# insert new row 3
ws.insert_rows(3)
```

```python
# verify insert
for row in ws.rows:
    for cell in row:
        print(cell.value)
```

```python
# save the file
wb.save('large_file_insert_rows.xlsx')
```

## Deleting rows and columns


To delete the columns `C:E`:

```python
from openpyxl import load_workbook

wb = load_workbook(filename='large_file_before.xlsx', read_only=False)
ws = wb['big_data']

for row in ws.rows:
    for cell in row:
        print(cell.value)
```

```python
# delete columns `C:E`
ws.delete_cols(3, 3)
```

```python
# verify insert
for row in ws.rows:
    for cell in row:
        print(cell.value)
```

```python
# save the file
wb.save('large_file_delete_cols.xlsx')
```

![](./large_file_delete_cols.png)

>  Note:
>
>   Openpyxl does not manage dependencies, such as formulae, tables, charts,
>   etc., when rows or columns are inserted or deleted. This is considered to
>   be out of scope for a library that focuses on managing the file format.
>   As a result, client code **must** implement the functionality required in
>   any particular use case.

## Moving ranges of cells


You can also move ranges of cells within a worksheet::

```python
from openpyxl import load_workbook

wb = load_workbook(filename='large_file_before.xlsx', read_only=False)
ws = wb['big_data']

for row in ws.rows:
    for cell in row:
        print(cell.value)
```

```python
# move cell rang `C3:E5` up one row, and right two columns
ws.move_range("C3:E5", rows=-1, cols=2)
```

```python
# verify insert
for row in ws.rows:
    for cell in row:
        print(cell.value)
```

```python
# save the file
wb.save('large_file_move_range.xlsx')
```

![](./large_file_move_range.png)

This will move the cells in the range `C3:E5` up one row, and right two
columns. The cells will overwrite any existing cells.

If cells contain formulae you can let openpyxl translate these for you, but
as this is not always what you want it is disabled by default. Also only the
formulae in the cells themselves will be translated. References to the cells
from other cells or defined names will not be updated; you can use the
`formula` translator to do this::

```python
from openpyxl import load_workbook

wb = load_workbook(filename='large_file_before.xlsx', read_only=False)
ws = wb['big_data']

for row in ws.rows:
    for cell in row:
        print(cell.value)
```

```python
# move and translate: cell rang `C4:D5` up one row, and right one column 
ws.move_range("C4:D5", rows=1, cols=1, translate=True)
```

```python
# verify insert
for row in ws.rows:
    for cell in row:
        print(cell.value)
```

```python
# save the file
wb.save('large_file_move_translate_range.xlsx')
```

![](./large_file_move_translate_range.png)

This will move the relative references in formulae in the range by one row and one column.
